```python
#Import Pandas for forther data transormation
import pandas as pd
```

```python
#Import our file
df = pd.read_csv('/media/vlad/Transcend/Microsoft VS Code/ready code/Data_science_projects/train.csv')
```

```python
#Apply two operators with the reasone to check our dataframe
df.columns()
df.head(10)
```

```python
#Receive info about our dataframe
df.info()
```

```python
#Lets detect how much zeros we have in our dataframe
df.isnull().sum()
```

```python
#Drop several columnes we dont need
df = df.drop(['Name','Ticket',"Cabin",'PassengerId'], axis= 1)
```

```python
#Our main seria will be applied and lets find out what we have
df['Sex'].value_counts()
```

```python
#binare coding for our columne df.sex 
sex_mapping = {'male':0,'female':1}
df['Sex'] = df.Sex.map(sex_mapping)
```

```python
df.head()
```

```python
df["Embarked"].value_counts()
```

```python
#One hot endocing, wectore of len_3
#S -> 001/C -> 100/Q ->010
# with the function get_dummies()
Embarked_dummies = pd.get_dummies(df["Embarked"], prefix= 'port', dummy_na= False)
```

```python
Embarked_dummies.head()
```

```python
#Compile our 2 DataFrames
df = pd.concat([df,Embarked_dummies], axis = 1)
```

```python
df = df.drop(["Embarked"], axis = 1)
```

```python
df
```

```python
#To form data for typical ML
X = df
X = X.drop(["Survived"], axis = 1)
y = df["Survived"]
```

```python
X.head()
```

```python
#Creation of deciosin tree
X.isnull().sum()
```

```python
X['Age'] = X["Age"].fillna(value= X["Age"].mean())
```

```python
X
```

```python
#As the first step we must train our model and than check it
import numpy as np 
from sklearn.model_selection import KFold, cross_val_score, train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.33, random_state = 42)  
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier(max_depth= 5)
model_tree.fit(X_train, y_train)
```

```python
#use .dot with the reasone to vizual the tree
from sklearn.tree import export_graphviz
export_graphviz(model_tree, feature_names= ['Pclassn','Sex','Age','SibSP','Parch','Fare','port_C','port_Q','port_S'], out_file = 'tree.dot', filled = True)
#with the reasone to vizual we need package pydot(pip intall pydot)
!dot -Tpng 'tree.dot' -o 'tree.png'
```

```python
#Tree was created now must be done prediction
y_predict = model_tree.predict(X_test)
```

```python
y_predict
```

```python
#apply metric accuracy_score in order to mentoring of efficinacy of the model
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test,y_predict))
print('Confiusion matrix: \n' ,confusion_matrix(y_test,y_predict))
print()
print('Accuracy:', accuracy_score(y_test,y_predict))
```

```python
Predict = pd.DataFrame(y_predict)
df = pd.concat([df,Predict], axis = 1) #Compile our DataFrames and serias with predict
```
